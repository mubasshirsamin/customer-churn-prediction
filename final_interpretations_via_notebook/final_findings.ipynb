## Coefficients are conditional on other variables
Logistic Regression is a model that does not describe a feature in isolation. It describes the effect of one feature while holding the others constant. This suggests some variables can behave differently in this model compared to their behavior in a simple graph. This suggests if a feature may appear positively related to churn, but after the model accounts for other features, the coefficients can shrink or even change sign.

In this dataset:
- Tenure
- Monthly Charges
- Total Charges

Overlaps with each other.

This is because total charges are naturally influenced by how long a customer has stayed and how much they pay each month. As a result, the model is trying to separate overlapping pieces of information across these three variables. This is why TotalCharges being positive and MonthlyCharges being negative may look counterintuitive. This does not necessarily mean the model is wrong. It means the model is distributing the effect across several correlated financial variables, so the coefficients are harder to interpret one by one.

## One-hot encoded variables are interpreted relative to a baseline category

Some of the categorical variables in the dataset, such as contract type and internet service, were converted into separate yes/no columns using one-hot encoding. Because of this, the model does not interpret each category completely on its own. Instead, it compares each shown category to one category that was left out as the reference group.

For example, the variables Contract_One year and Contract_Two year are most likely being compared to customers on month-to-month contracts, which acts as the baseline. So when these contract coefficients are negative, it means customers on one-year or two-year contracts are less likely to churn compared with month-to-month customers.

In the same way, InternetService_Fiber optic is being interpreted relative to the internet service category that was omitted during encoding. So the coefficient tells us whether fiber optic customers are more or less likely to churn compared with that reference internet group, not in a completely separate absolute sense.

So the main idea is this: these coefficients are showing how a category behaves compared to a baseline group, not how it behaves in isolation.
