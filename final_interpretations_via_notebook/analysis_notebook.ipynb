# Customer Churn Prediction - Exploratory Data Analysis

This notebook analyses the telecom churn dataset to identify customer patterns, service-related trends, and the possible driving factors of churn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("output/cleaned_telco_churn.csv")

print("Loaded Dataset successfully.")
print("Shape:", df.shape)
df.head()
```

# Overview of the Dataset

The cleaned dataset contains 7043 customer records and 21 variables.  
The target variable is `Churn`, where:
- `1` means the customer churned
- `0` means the customer stayed

The target variable is analysed with its possible driving factors to make insightful
interpretations on the reasons for customers leaving the company.

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Churn")
plt.title("Overall Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.xticks([0, 1], ["No", "Yes"])
plt.tight_layout()
plt.show()
```

For the telecom company in this dataset, most customers stayed with the service, while only a small portion decided to leave. Approximately, 26.5% of the total customers churned,suggesting that this case is not a rare event but a meaningful issue for the company.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Contract", hue="Churn")
plt.title("Churn by contract type")
plt.xlabel("Contract type")
plt.ylabel("Customer number")
plt.tight_layout()
plt.show()
```

The above bar graph illustrates that customers enrolled on a month-month contract appear to churn much more often than customers enrolled on yearly contracts. This suggests shorter contract commitment will associate the company with a higher churn risk for the company's customer base.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Churn", y="MonthlyCharges")
plt.title("Monthly Charges by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges($)")
plt.xticks([0, 1], ["No", "Yes"])
plt.tight_layout()
plt.show()
```

The median for churned customers is higher than non-churned customers suggesting that customers who churned appear to have higher monthly charges on average than customers whon stayed. These customers paying more each month may feel that the service is too expensive compared to the value they recieve, making them likely to leave. The company may need to improve value communication, provide discounts or plan a better monthly payment structure to prevent this.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="InternetService", hue="Churn")
plt.title("Churn by Internet Service")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

From the above graph, it can be interpretated that customers paying for fiber optic service may expect better performance and feel they are paying more than the value they recieve. As Fiber optic is seen as a premium service if those expectations are not met they are more likely to leave as illustrated in the graph. DSL customers on the other hand might have lower monthly cost with moderate service expectations making them less likely to churn compared to those paying for Fiber optic internet service.

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="PaymentMethod", hue="Churn")
plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

The above graph illustrates that customers making payments via electronic check are more likely to churn. If the payment process is manual and is slower and less familiar to the customer, this can be a key factor on making them leave the service. More billing frictions can cause more inconvenice for customers. On the other hand, automatic payment services such as payment via credit card can promote retention in the company's consumer base as it does generate billing frictions and is convenient for most customers.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="TechSupport", hue="Churn")
plt.title("Churn by Tech Support")
plt.xlabel("Tech Support")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

The above graph illustrates that customer who get faster and technical support are less likely to leave due to improved customer experience. The company could bundle tech support in internet packages which will in turn improve retention for customers using internet services, as customers without readily available assistance may get frustrated and are more likely to churn.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="OnlineSecurity", hue="Churn")
plt.title("Churn by Online Security")
plt.xlabel("Online Security")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

The bar graph shows customers who purchase the add-on service of online security are more likely to stay within the company potentially linked to customer satisfaction and its perceived value. Company can impose marketing strategies to make this feature more affordable and avaialable for customers to improve retention

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x="PaperlessBilling", hue="Churn")
plt.title("Churn by Paperless Billing")
plt.xlabel("Paperless Billing")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

The higher churn observed among customers using paperless billing suggests that this billing format may be associated more with customer behaviour, but the relationship is likely indirect. This variable is likely to be a secondary indicator and may reflect the overlap with stronger factors such as Tech support, contract type and payment method. 

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Churn", y="tenure")
plt.title("Tenure by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Tenure (months)")
plt.xticks([0, 1], ["No", "Yes"])
plt.tight_layout()
plt.show()
```

The above box-plot explains that customers staying with the company for a longer time is less likely to churn compared to customers who have a lower tenure, as the median for churned customers is concentrated in the lower tenure range. This factor is not considered to be as strong other factors such as payment method and monthly charges in driving the level of churned customers within the company.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Churn", y="TotalCharges")
plt.title("Total Charges by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Total Charges($)")
plt.xticks([0, 1], ["No", "Yes"])
plt.tight_layout()
plt.show()
```

This above box-plot demonstrated that customers who did not churn have higher total charges piled up in their duration of purchasing services within the company, while in the other hand customers who have churned have a lower level of total charges value. Although, there are some cases where customers with a high total charges value have churned which are represented as outliers in the second box-plot.
